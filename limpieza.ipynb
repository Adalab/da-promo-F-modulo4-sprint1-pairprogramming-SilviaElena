```python
whr_copy=whr.copy()
```

```python
whr_copy.columns
```

```python
imputer_iterative = IterativeImputer(max_iter =20, random_state=42)
imputer_iterative_imputado = imputer_iterative.fit_transform(whr_copy[['Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Positive affect', 'Negative affect']])
imputer_iterative_imputado
```

```python
#KNN IMPUTER
imputer_knn = KNNImputer(n_neighbors=5)
imputer_knn_imputado = imputer_knn.fit_transform(whr_copy[['Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Positive affect', 'Negative affect']])
imputer_knn_imputado
```

```python
whr_copy[['Log GDP per capita_knn',
       'Social support_knn', 'Healthy life expectancy at birth_knn',
       'Freedom to make life choices_knn', 'Generosity_knn',
       'Perceptions of corruption_knn', 'Positive affect_knn', 'Negative affect_knn']] = imputer_knn_imputado
```

```python
whr_copy[['Log GDP per capita_iterative',
       'Social support_iterative', 'Healthy life expectancy at birth_iterative',
       'Freedom to make life choices_iterative', 'Generosity_iterative',
       'Perceptions of corruption_iterative', 'Positive affect_iterative', 'Negative affect_iterative']] = imputer_iterative_imputado
```

```python
#Comparamos las estadísticas generadas de por el iterative y el knn para decidir cuál coger (aquel cuya media y mediana sea más parecida a la original- En la columna con valores atípicos cogemos aquel cuya mediana sea más parecida a la original)
whr_copy.describe()[['Log GDP per capita', 'Log GDP per capita_knn', 'Log GDP per capita_iterative','Social support','Social support_knn','Social support_iterative','Healthy life expectancy at birth','Healthy life expectancy at birth_knn','Healthy life expectancy at birth_iterative','Freedom to make life choices','Freedom to make life choices_knn','Freedom to make life choices_iterative','Generosity','Generosity_knn','Generosity_iterative','Perceptions of corruption','Perceptions of corruption_knn','Perceptions of corruption_iterative','Positive affect','Positive affect_knn','Positive affect_iterative','Negative affect','Negative affect_knn','Negative affect_iterative']].T
```
